## Scatter Plot

```python
x = np.array([5,7,8,7,2,10,3,4,5,6,7,12])
y = np.array([90,85,86,74,96,45,74,69,68,56,98,74])

plt.scatter(x,y,c='red')
plt.show()
```

```python
x = np.array([5,7,8,7,2,10,3,4,5,6,7,12])
y = np.array([90,85,86,74,96,45,74,69,68,56,98,74])
colors = np.array([0,5,10,15,20,25,30,35,40,45,50,55])
size = np.array([100,200,300,400,500,600,550,450,350,250,150,50])

plt.scatter(x,y,cmap = 'rainbow_r', c = colors, s=size)
plt.colorbar()
plt.show()
```

```python
x = np.random.randint(100,size=(100))
y = np.random.randint(100,size=(100))
colors = np.random.randint(100,size=(100))
size = 10*np.random.randint(100,size=(100))

plt.scatter(x,y,c=colors,s=size,cmap='nipy_spectral', alpha=0.5)
plt.colorbar()
plt.show()
```

## Subplot

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(15,15))
plt.subplot(3,2,1)
x = np.array([1,2,3,4,5])
y = np.array([3,1,7,6,4])
plt.plot(x,y,marker='*')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Line chart')

plt.subplot(3,2,2)
plt.bar(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Bar Graph')

plt.subplot(3,2,3)
plt.pie(y,autopct='%0.2f%%')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Pie Chart')

plt.subplot(3,2,4)
plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot')

plt.subplot(3,2,5)
plt.fill_between(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Area Plot')

plt.subplot(3,2,6)
plt.stackplot(x,y,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Stack Plot')
plt.suptitle("Subplot")
plt.savefig('image.png')
plt.savefig('image.svg')
plt.show()
```

```python
Scores= np.array([[13, 10, 9, 33], 
[63, 46, 90, 42], 
[39, 76, 13, 29], 
[82, 9, 29, 78], 
[67, 61, 59, 36]])
Match_5= np.array([41, 87, 72, 36, 92]).reshape(5,1)
Scores = np.concatenate((Scores,Match_5),axis=1)
Batsman_6= np.array([77, 83, 98, 95, 89]).reshape(1,5)
Batsman_7= np.array([92, 71, 52, 61, 53]).reshape(1,5)
Scores = np.concatenate((Scores,Batsman_6,Batsman_7),axis=0)
L = []
for i in Scores:
    L.append(sum(i))
L = np.array(L).reshape(7,1)
Scores = np.concatenate((Scores,L),axis=1)
print(Scores)

total_scores = np.sort(Scores[:,-1])
players = [1,2,3,4,5,6,7]
plt.plot(players,total_scores,ls='--',color='k')
plt.xlabel("No. of Batsman")
plt.ylabel("Scores")
plt.title("Leaderboard",fontdict={'weight':'bold'})
plt.show()

bat1 = Scores[0,:-1]
bat2 = Scores[1,:-1]
match = np.array([1,2,3,4,5])
plt.bar(match,bat2,color='darkred',label='Batsman-2')
plt.bar(match,bat1,color='purple',label='Batsman-1')
plt.legend()
plt.show()

bat1 = Scores[0,:-1]
bat2 = Scores[1,:-1]
match = np.array([1,2,3,4,5])
plt.bar(match,bat2,color='darkred',label='Batsman-2',width=0.35)
plt.bar(match+0.35,bat1,color='purple',label='Batsman-1',width=0.35)
plt.legend()
plt.show()

plt.pie(Scores[:,-1],explode=[0.1]*7,autopct='%0.1f%%',labels=[1,2,3,4,5,6,7])
plt.legend()
plt.show()
```
