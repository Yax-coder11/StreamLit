## Multilevel Inheritance

```python
class Father:
    def __init__(self):
        print("Inside Father class")
        self.a = 5
    def disp(self):
        print("Father instance method")
        print(self.a)
class Child(Father):
    def __init__(self):
        super().__init__()
        print("Inside Child class")
    def show(self):
        print("Child instance method")
class GrandChild(Child):
    def __init__(self):
        super().__init__()
        print("Inside Grand child class")
    def show(self):
        print("Grand child instance method")
        super().show()
g = GrandChild()
g.show()
g.disp()
```

```python
from abc import ABC, abstractmethod
import math
class Shape(ABC):
    @abstractmethod
    def calculate_area(self):
        pass
class Rectangle(Shape):
    def __init__(self,l,b):
        self.length = l
        self.breadth = b
    def calculate_area(self):
        return self.length * self.breadth
    def __gt__(self,other):
        if self.calculate_area() > other.calculate_area():
            return "First Rectangle has greater area"
        else:
            return "Second Rectangle has greater area"
class Circle(Shape):
    def __init__(self,r):
        self.radius = r
    def calculate_area(self):
        return math.pi*self.radius*self.radius
c = Circle(6)
r1 = Rectangle(3,9)
r2 = Rectangle(4,5)
print(c.calculate_area())
print(r1>r2)
print(Circle.mro())
```

```python
from abc import ABC, abstractmethod
class Employee(ABC):
    def get_empdetails(self):
        self.empid = int(input("Enter employee id: "))
        self.name = input("Enter Name: ")
        self.salary = int(input("Enter basic salary: "))
    def print_empdetails(self):
        print(f"Employee id: {self.empid}")
        print(f"Name: {self.name}")
        print(f"Basic Salary: {self.salary}")
    def get_salary(self):
        return self.salary
    @abstractmethod
    def emp_id(self):
        pass
class Perks(Employee):
    def calculate_perks(self):
        self.da = 0.35*self.salary
        self.hra = 0.17*self.salary
        self.pf = 0.12*self.salary
    def print_perks(self):
        print(f"DA: {self.da}")
        print(f"HRA: {self.hra}")
        print(f"PF: {self.pf}")
        print(f"Total Perks: {self.da+self.hra-self.pf}")
class NetSalary(Perks):
    def netsalary(self):
        self.net_salary = self.salary+self.da+self.hra-self.pf
    def print_netsalary(self):
        self.print_empdetails()
        self.print_perks()
        print(f"Total Salary: {self.net_salary}")
    def emp_id(self):
        return self.empid
e = NetSalary()
e.get_empdetails()
e.calculate_perks()
e.netsalary()
e.print_netsalary()
```

```python
import numpy as np
class Matrix:
    def __init__(self,r,c):
        self.rows = r
        self.columns = c
        self.mat = np.zeros((r,c),dtype=int)
    def get_rows(self):
        return self.rows
    def get_columns(self):
        return self.columns
    def set_elements(self):
        for i in range(self.rows):
            for j in range(self.columns):
                self.mat[i][j] = int(input(f"Enter element on {i} row, {j} column"))
    def __add__(self,other):
        if self.mat.shape == other.mat.shape:
            return self.mat + other.mat
        else:
            return "Matrices can't be added"
    def __mul__(self,other):
        if self.columns == other.rows:
            return np.dot(self.mat,other.mat)
        else:
            return "Matrices can't be Multiplied"
m1 = Matrix(2,2)
m1.set_elements()
m2 = Matrix(2,2)
m2.set_elements()
print(m1+m2)
print(m1*m2)
```

```python
import numpy as np
class BaseSalesData:
    sales = np.array([ 
[[120,10], [140,12], [130,11], [150,13], [135,10]], 
[[80,15],  [90,18],  [85,16],  [88,17],  [92,19]], 
[[200,8],  [210,9],  [205,10], [215,8],  [220,7]], 
[[60,20],  [65,22],  [70,21],  [68,23],  [72,24]], 
[[150,12], [155,13], [148,14], [160,15], [158,12]], 
[[230,6],  [240,7],  [235,6],  [245,8],  [250,7]] 
]) 
    def __init__(self):
        self.sales = BaseSalesData.sales
        self.products = self.sales.shape[0]
        self.regions = self.sales.shape[1]

class SalesStabilityAnalysis(BaseSalesData):
    
    def net_units(self):
        net=[]
        for p in range(self.products):
            row = []
            for r in range(self.regions):
                row.append(self.sales[p][r][0] - self.sales[p][r][1])
            net.append(row)
        return np.array(net)
    def risk_products(self):
        risky=[]
        for p in range(self.products):
            for r in range(self.regions):
                if self.sales[p][r][1]>15:
                    risky.append(p)
                    break
        return risky
    def regional_reliability(self):
        net = self.net_units()
        avg = []
        for r in range(self.regions):
            s = 0
            for p in range(self.products):
                s += net[p][r]
            avg.append(s / self.products)
        return avg
    def most_stable_product(self):
        net = self.net_units()
        min_values=[]
        for p in range(self.products):
            min_val = net[p][0]
            for r in range(1,self.regions):
                if net[p][r]<min_val:
                    min_val = net[p][r]
            min_values.append(min_val)
        return np.argmax(min_values)
    def normalize_returns(self):
        for p in range(self.products):
            for r in range(self.regions):
                if 14<=self.sales[p][r][1]<=16:
                    self.sales[p][r][1] = 14
        return self.net_units()
    def stability_index(self):
        net = self.net_units()
        index = []
        for p in range(self.products):
            total = 0
            for r in range(self.regions):
                total += net[p][r]
            avg = total/self.regions
            index.append((avg/200)*100)
        return index
    def forecast_readiness(self):
        index = self.stability_index()
        status = []
        for v in index:
            if v>=75:
                status.append('High')
            elif v>=60:
                status.append('Moderate')
            elif v>=50:
                status.append('Low')
            else:
                status.append('Unstable')
        return status
    def planning_priority(self):
        index = self.stability_index()
        temp = list(enumerate(index))
        temp.sort(key=lambda x:x[1],reverse=True)
        order = []
        for i in temp:
            order.append(i[0])
        return order
sa = SalesStabilityAnalysis()
print("Net Units:\n", sa.net_units())
print("Risk Products:", sa.risk_products())
print("Regional Reliability:", sa.regional_reliability()) 
print("Most Stable Product ID:", sa.most_stable_product()) 
print("After Return Normaliza on:\n", sa.normalize_returns())
print("Stability Index:", sa.stability_index()) 
print("Forecast Readiness:", sa.forecast_readiness()) 
print("Planning Priority:", sa.planning_priority()) 
```
